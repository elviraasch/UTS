```python
import pandas as pd

wine_df = pd.read_csv('winequality-red.csv')

winered = ['alcohol', 'pH', 'residual sugar', 'citric acid']

for col in winered:
    median_val = wine_df[col].median()
    df_grouped = wine_df.groupby(wine_df[col].apply(lambda x: 'high' if x >= median_val else 'low'))

    print(df_grouped['quality'].mean())
    print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print(wine_df.info())
print(wine_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7, 4))
sns.countplot(data=wine_df, x='quality', palette='coolwarm')

plt.title('Distribusi Peringkat Kualitas Anggur (Quality Rating Score)', fontsize=12, pad=10)
plt.xlabel('Skor Kualitas')
plt.ylabel('Jumlah Sampel (Count)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9, 7))

matriks_korelasi = wine_df.corr()

sns.heatmap(matriks_korelasi, annot=True, cmap='RdBu_r', fmt='.3f', linewidths=0.5, vmin=-1, vmax=1)

plt.title('Matriks Korelasi Hubungan Sifat Fisikokimia dengan Kualitas Anggur')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(15, 11))

# Alcohol vs Quality
sns.boxplot(ax=axes[0, 0], data=wine_df, x='quality', y='alcohol', palette='Set2')

# pH vs Quality
sns.boxplot(ax=axes[0, 1], data=wine_df, x='quality', y='pH', palette='Set2')

# Residual Sugar vs Quality
sns.boxplot(ax=axes[1, 0], data=wine_df, x='quality', y='residual sugar', palette='Set2')

# Citric Acid vs Quality
sns.boxplot(ax=axes[1, 1], data=wine_df, x='quality', y='citric acid', palette='Set2')

plt.tight_layout(pad=3.0)
plt.show()
```
